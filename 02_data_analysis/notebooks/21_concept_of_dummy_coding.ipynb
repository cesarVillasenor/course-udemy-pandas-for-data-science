# 21 Concept of Dummy Coding

```python
import pandas as pd
```

```python
titanic = pd.read_csv('http://bit.ly/kaggletrain')
titanic.to_csv('../data/kaggletrain.csv')
titanic.head()
```

```python
titanic.Sex.map({'male': 1, 'female': 0})
```

```python
titanic['IsMale'] = titanic.Sex.map({'male': 1, 'female': 0})
```

```python
titanic.head()
```

```python
titanic.Embarked.unique()
```

```python
pd.get_dummies(titanic.Embarked)
```

```python
embark = pd.get_dummies(titanic.Embarked, prefix='dummy')
```

```python
pd.concat([titanic, embark], axis=1).head()
```
