# 26 Example on Copy Warning

```python
import pandas as pd
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

```python
drinks_wine = drinks.loc[drinks.wine_servings == 0, :]
```

```python
drinks_wine.head()
```

```python
import numpy as np
```

```python
drinks_wine.loc[:, 'wine_servings'] = np.nan
```

```python
drinks_wine = drinks.loc[drinks.wine_servings == 0, :].copy()
```

```python
drinks_wine.loc[:, 'wine_servings'] = np.nan
```
