# 07 Sorting Data Frame

```python
import pandas as pd
```

```python
data = pd.read_csv('http://bit.ly/imdbratings')
# data.to_csv('../data/imdbratings.csv')
data.head()
```

```python
data['duration'].sort_values(ascending=False)
```

```python
data.sort_values('duration', ascending=False, inplace=True)
data.head()
```

```python
data.sort_values(['genre', 'duration'], ascending=False, inplace=True)
data.head()
```
