# 13 Pivot Table Method

```python
import pandas as pd
```

```python
data =  pd.read_csv('../data/Temperature-2.csv')
```

```python
data
```

```python
# error since theya duplicate values
data.pivot(index='Day', columns='City')
```

```python
data.pivot_table(index='Day', columns='City')
```

```python
data.pivot_table(index='Day', columns='City', aggfunc=min)
```

```python
data.pivot_table(index='Day', columns='City', aggfunc=max, margins=True)
```
