# 03 Shaping a Data Frame

```python
import pandas as pd
```

```python
df1 = pd.DataFrame([['TN', 'Chennai', 'Tamil'], ['Telangana', 'Hyderabad', 'Telugu'], ['Karnataka', 'Bangalore', 'Kannada']], columns=['State', 'Capital', 'Languages'])
```

```python
df1
```

```python
df2 = pd.melt(df1, id_vars='State')
```

```python
df2
```

```python
df2 = pd.DataFrame({'days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], 'temperature': [30,32,31,34,32,34,35], 'humidity': [30,32,31,34,32,34,35]})
```

```python
pd.melt(df2, id_vars='days')
```

```python
pd.melt(df2, id_vars='days', var_name='Metric', value_name='Humidity/Temperature Value')
```
