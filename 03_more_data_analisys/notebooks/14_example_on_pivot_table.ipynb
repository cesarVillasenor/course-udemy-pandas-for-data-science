# 14 Example on Pivto Table 

```python
import pandas as pd
```

```python
data =  pd.read_csv('../data/Temperature-3.csv')
```

```python
data
```

```python
data.pivot_table(index='Day', columns='City')
```

```python
type(data.Day[0])
```

```python
data.Day = pd.to_datetime(data.Day)
type(data.Day[0])
```

```python
data.pivot_table(index=pd.Grouper(key='Day', freq='W'), columns='City')
```

```python
data.pivot_table(index=pd.Grouper(key='Day', freq='W'), columns='City', aggfunc=max)
```
