# 07 Stacking and Unstacking

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('../data/products.xlsx', header=[0,1], index_col=[0])
```

```python
data.head()
```

```python
data.stack(level=0)
```

```python
data.stack(level=[0,1])
```

```python
data_stack = data.stack()
data_stack
```

```python
data_stack = data_stack.stack()
data_stack
```

```python
data_stack.unstack()
```

```python
data_stack.unstack(level=0)
```

```python
data_stack.unstack(level=[0,2])
```
