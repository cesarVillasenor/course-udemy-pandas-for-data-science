# 12 Functions of Pivot

```python
import pandas as pd
```

```python
data =  pd.read_csv('../data/Temperature.csv')
```

```python
data
```

```python
data.pivot(index='Day', columns='City', values='Precipation')
```

```python
data.pivot(index='Day', columns='City', values='Wind Speed')
```

```python
data.pivot(index='Day', columns='City', values=['Wind Speed', 'Precipation'])
```
